导入卧龙区数据

```python
wolong_data = pd.read_csv('wolong_income.csv')
wolong_data["比对结果"] = "无匹配数据"
wolong_data["匹配次数"] = 0
wolong_data
```

删除无用数据

```python
#wolong_data_2 = wolong_data.iloc[:,1:5]
#wolong_data_2
```

删除表

```python
#wolong_data_2.drop(wolong_data_2.columns[[1,2]],axis=1,inplace=True)
#wolong_data_2
```

```python
for index_wl, row_wl in wolong_data.iterrows():
    name_wl = row_wl["申请人"]
    amount_wl = row_wl["土地出让金金额"]
    mof_data_tmp = mof_data[mof_data["缴款单位"]==name_wl]
    #print("卧龙区数据：索引%s 姓名 %s 金额 %s" % (index_wl, name_wl,amount_wl))
    
    count = 0
    for index_mof, row_mof in mof_data_tmp.iterrows():
        name_mof = row_mof["缴款单位"]
        amount_mof = row_mof["金额"]
        #print("财政局数据：索引%s 姓名 %s 金额 %s" % (index_mof, name_mof,amount_mof))

        if(amount_wl == amount_mof):
            count+=1
            #wl = wolong_data.iloc[index_wl, :]
            #mof = mof_data.iloc[index_mof, :]
            wolong_data.loc[index_wl,'比对结果']="匹配"
            wolong_data.loc[index_wl,'匹配次数']=count
```

风帆小区数据

```python
fengfan_data = pd.read_csv('fengfan_income.csv', thousands=',')
fengfan_data["比对结果"] = "无匹配数据"
fengfan_data["匹配次数"] = 0
```

```python
for index_fengfan, row_fengfan in fengfan_data.iterrows():
    name_fengfan = row_fengfan["申请人"]
    amount_fengfan = row_fengfan["补缴土地价款"]
    mof_data_tmp = mof_data[mof_data["缴款单位"]==name_fengfan]
    #print("卧龙区数据：索引%s 姓名 %s 金额 %s" % (index_wl, name_wl,amount_wl))
    
    count = 0
    for index_mof, row_mof in mof_data_tmp.iterrows():
        name_mof = row_mof["缴款单位"]
        amount_mof = row_mof["金额"]
        #print("财政局数据：索引%s 姓名 %s 金额 %s" % (index_mof, name_mof,amount_mof))
        
        if(amount_fengfan == amount_mof):
            count+=1
            fengfan_data.loc[index_fengfan,'比对结果']="匹配"
            fengfan_data.loc[index_fengfan,'匹配次数']=count
```

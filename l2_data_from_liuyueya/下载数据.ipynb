```python
def tsDateToDate(day):
    tsDate = datetime.datetime.strptime(day, '%Y%m%d')
    return tsDate.strftime("%Y-%m-%d")

tsDateToDate("20180117")
```

```python
def dateToTsDate(day):
    d = datetime.datetime.strptime(day, '%Y-%m-%d')
    return d.strftime("%Y%m%d");


dateToTsDate("2018-01-17")
```

```python
def todayToTsDate():
	return time.strftime("%Y%m%d", time.localtime());

todayToTsDate()
```

```python
def getTradeCal():
    pro = ts.pro_api('d94b8d1af9f3110dca7acf2e85b4bf10b7d33de74491de8f671c4b8b')
    trade_cal_data = pro.trade_cal(exchange='', start_date='20200601', end_date=todayToTsDate(), is_open=1)
    return trade_cal_data

getTradeCal()
```

```python
trade_cal = getTradeCal()
```

```python
from urllib import request
from urllib import error
```

```python
for index, row in trade_cal.iterrows():
    d = tsDateToDate(row["cal_date"])
    url = "https://ohlc1s.s3.cn-east-2.jdcloud-oss.com/"
    file = "%s.7z" % d
    
    #print(url+file)
    
    req = request.Request(url+file)
    try:
        response=request.urlopen(req)
    except error.URLError as e:
        print("文件不存在,地址%s" % url+file)
    else:
        '''
        print("文件存在,地址%s" % url+file)
        '''
        with open(file, "wb") as code:
            code.write(response.read())
        response.close()
    finally:
        pass
```
